```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = pd.read_csv('C:/Users/Mosi/Python Projects/TA Pythone/Stars/Stars.csv')
data
```

```python
x=np.arange(0,240,1)
t=data['Temperature'].values
l=data['L'].values
r=data['R'].values
```

```python
plt.figure(figsize=(9,5),dpi=180)
plt.plot(x,t)
plt.title('Temperature of Stars Plot')
plt.xlabel('Number of Type of Stars')
plt.ylabel('Temperature')
plt.show()
```

```python
plt.figure(figsize=(9,5),dpi=180)
plt.hist(t)
plt.title('Histogram of Stars Temperature')
plt.xlabel('Temperature')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
print('Variance: ',np.var(t),'Standard Deviation: ',np.std(t),'Mean: ',np.mean(t),'Median: ',np.median(t))
```

```python
plt.figure(figsize=(9,5),dpi=180)
plt.plot(x,l)
plt.title('Luminosity of Stars Plot')
plt.xlabel('Number of Type of Stars')
plt.ylabel('L/L0')
plt.show()
```

```python
plt.figure(figsize=(9,5),dpi=180)
plt.hist(l)
plt.title('Histogram of Stars Luminosity')
plt.xlabel('L/L0')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
print('Variance: ',np.var(l),'Standard Deviation: ',np.std(l),'Mean: ',np.mean(l),'Median: ',np.median(l))
```

```python
print('Covariance of Luminosity and Temprature: ')
print(np.cov(l,t))
```

```python
from scipy.stats import pearsonr
p_coeff, p_value=pearsonr(l,t)
```
