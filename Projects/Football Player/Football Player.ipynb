```python
#this player is playing for the left side...
import matplotlib.pyplot as plt
class footballplayer:
  def __init__(self,name,x,y,energy,ballx,bally):
    self.name=name
    self.x=x
    self.y=y
    self.energy=energy
    self.ballx=ballx
    self.bally=bally
    
  #movment function
  def move_x_up(self):
    self.x+=1
    self.ballx+=1
  def move_x_down(self):
    self.x-=1
    self.ballx-=1
  def move_y_up(self):
    self.y+=1
    self.bally+=1
  def move_y_down(self):
    self.y-=1
    self.bally-=1
    
  #energy function
  def rest(self):
    if self.energy<100 :
      self.energy=self.energy+5
    else :
      pass
  def run(self):
    self.energy=self.energy-5
    
  #dribbble movment function
  def dribbble_up(self):
    self.x=1+2
    self.ballx=1+2
    self.y=1+2
    self.bally=1+2
  def dribbble_down(self):
    self.x=1+2
    self.ballx=1+2
    self.y=1-2
    self.ballx=1-2
    
  #dive movment function
  def dive_right(self):
    self.x=1+3
  def dive_left(self):
    self.x=1-3
  def dive_up(self):
    self.y=1+3
  def dive_down(self):
    self.y=1-3

  #shoot function
  def shoot_x_up(self):
    self.ballx=1+5
  def shoot_x_down(self):
    self.ballx=1-5
  def shoot_y_up(self):
    self.bally=1+5
  def shoot_y_down(self):
    self.bally=1-5
    
  #information function
  def get_info(self):
    print('name :',self.name,'|x :',self.x,'|y :',self.y,'|energy :',self.energy)
    plt.scatter(self.x,self.y)
    plt.scatter(self.ballx,self.bally)
    plt.grid()
    plt.xlabel('x position')
    plt.ylabel('y position')
    plt.title('position of football player with ball')
    plt.show()
```

```python
RealMadrid=footballplayer('Mosi',0,0,100,0,0)
RealMadrid.get_info()
```

```python
RealMadrid.rest()
RealMadrid.get_info()
```
