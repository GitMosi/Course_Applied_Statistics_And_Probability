```python
from math import factorial as fa
def sin(x):
    s=0
    for i in range(0,10):
        s+=(((((-1)**i))*(x**((2*i)+1)))/(fa((2*i)+1)))
    return s
```

```python
import numpy as np 
x=np.linspace(-8,8,160)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,5),dpi=180)
plt.plot(x,sin(x))
plt.scatter(x,sin(x))
plt.xlabel('x')
plt.ylabel('Sin(x)')
plt.title('Sine Taylor Series Plotting')
plt.show()
```
