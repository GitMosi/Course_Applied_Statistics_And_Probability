```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
```

```python
data = pd.read_csv('C:/Nezzzz/Python/TA/BTC/BTC.csv')
data
```

```python
x=np.arange(0,3705,1)
d=data['date'].values
o=data['1. open'].values
h=data['2. high'].values
l=data['3. low'].values
c=data['4. close'].values
v=data['5. volume'].values
```

```python
plt.figure(figsize=(6,5),dpi=80)
plt.plot(x,o)
plt.title('BTC Open Prices Plot')
plt.xlabel('Number of Day')
plt.ylabel('Open Price')
plt.show()
```

```python
plt.figure(figsize=(6,5),dpi=80)
plt.hist(o)
plt.title('Histogram of BTC Open Prices')
plt.xlabel('Open Price')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
print('Variance: ',np.var(o),'Standard Deviation ',np.std(o),'Mean: ',np.mean(o),'Median: ',np.median(o))
```
